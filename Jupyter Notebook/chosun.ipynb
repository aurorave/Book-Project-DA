```python
df_result = pd.DataFrame()

for i in stoplist:
    result = data[data['title'].str.contains(i)]
    df_result = pd.concat([df_result,result])
    print(df_result)
```

```python
# 확인용! (위의 137091 - 134869 = 2222)
df_test = df_result.drop_duplicates('title')
df_test.shape
```

```python
data = pd.merge(data, df_result, how='outer', indicator=True)
data
```

```python
data = data.query('_merge == "left_only"').drop(columns=['_merge'])
data

# 865633 - 136958 = 728675 여야 함 (약 2,215개 잉여: 최초 data에서 stoplist 해당 단어 포함 타이틀 존재)
```

```python
data['title'] = [re.sub('[^ \가-힣]', '', s) for s in data['title']]
data

# 최종 가용데이터 : 730764 개
```
